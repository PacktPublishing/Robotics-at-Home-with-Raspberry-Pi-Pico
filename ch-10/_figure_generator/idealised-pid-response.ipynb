```python
from matplotlib import pyplot as plt
import numpy as np
```

```python
# create a time linspace - up to 8 pi
t = np.linspace(0, 8 * np.pi, 1000)
# use this to make the idealised pid response graph, with a little overshoot and undershoot
error = np.sin(t) * 0.125
for n in range(1, 63, 2):
  error = np.sin(t / n)+ error
plt.plot(t, error)
plt.show()
```
